## Boosting

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=7, n_features=2, n_classes=2, n_clusters_per_class=2, class_sep=1.2, random_state=0, n_redundant = 0)
X = np.round(X, 1)
plt.scatter(X[:, 0], X[:, 1], c = y, cmap = 'bwr')
```

```python
y
```

```python
pred = y.copy()
pred[0] = 0
pred[1] = 1
pred
```

```python
w = np.ones(len(y))
```

```python
acc_w = ((y == pred)*w).sum()/w.sum()
acc_w
```

```python
5/7
```

```python
alpha = np.log(acc_w/(1-acc_w))
alpha
```

```python
np.log(2.5)
```

```python
x = np.linspace(1e-3, 1-1e-3, 100)
imp = np.log(x/(1-x))
plt.plot(x, imp)
plt.xlabel('Acc')
plt.ylabel('Importance of say')
plt.savefig('images/imp_of_say.pdf', bbox_inches = 'tight')
```

```python
w_new = w*np.exp((y != pred)*alpha)
w_new
```
